```python
sns.lineplot(x="size", y="total_bill", data=df, hue="sex", style="sex", palette="Accent",markers=["o","<"], legend="brief")
```

```python
sns.lineplot(x="size", y="total_bill", data=df, 
             hue="sex", style="sex", palette="Accent",
             dashes=False,markers=["o","<"], legend="brief")
plt.title=("Line Plot") #,fontsize = 20
plt.xlabel("size")#,fontsize = 15)
plt.ylabel("total_bill")#,fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style='darkgrid')
sns.lineplot(x="size", y="total_bill", data=df, 
             hue="sex", style="sex", palette="Accent",
             dashes=False,markers=["o","<"], legend="brief")
plt.title=("Line Plot") #,fontsize = 20
plt.xlabel("size")#,fontsize = 15)
plt.ylabel("total_bill")#,fontsize = 15)
plt.show()
```

```python
sns.set(style='darkgrid')
sns.lineplot(x="size", y="total_bill", data=df, 
             hue="day", style="day", palette="Accent",
             dashes=False,markers=["o","<"], legend="brief")
plt.title=("Line Plot") #,fontsize = 20
plt.xlabel("size")#,fontsize = 15)
plt.ylabel("total_bill")#,fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()

sns.distplot(df["total_bill"], hist_kws={'color': "#DC143C", 'edgecolor':"#aaff00",'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6}) #alpha is to give the shade of the color
```

```python
plt.figure(figsize=(16,9))
sns.set()

sns.distplot(df["total_bill"], 
             hist_kws={'color': "#DC143C", 'edgecolor':"#aaff00",'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6},
             kde_kws={'color': "#8e00ce", 'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6},# in kde_kws we cannot use edgecolor and kde_kws is for the line plot in the graph
             rug=True,
             rug_kws={'color': "#0426d0", 'edgecolor':"#aaff00",'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6})
```

```python
plt.figure(figsize=(16,9))
sns.set()

sns.distplot(df["total_bill"], 
             hist_kws={'color': "#DC143C", 'edgecolor':"#aaff00",'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6},
             kde =False,
             fit= norm,
             fit_kws={'color': "#8e00ce", 'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6},# while fit_kws we have to use kde=False  and by this we get the normalize curve in the graph nd it will be same as the kde argument
             rug=True,
             rug_kws={'color': "#0426d0", 'edgecolor':"#aaff00",'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6})# we can increase or decrease the value of alpha for the transparency
```

```python
plt.figure(figsize=(16,9))
sns.set()

sns.distplot(df["total_bill"], 
             hist_kws={'color': "#DC143C", 'edgecolor':"#aaff00",'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6},
             kde =False,
             fit= norm,
             fit_kws={'color': "#8e00ce", 'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6},# while fit_kws we have to use kde=False  and by this we get the normalize curve in the graph nd it will be same as the kde argument
             rug=True,
             rug_kws={'color': "#0426d0", 'edgecolor':"#aaff00",'linewidth':5, 'linestyle':'--',
                                        'alpha' :0.6},
            label ="TB")

plt.title("Histogram of Restorant Total Bill", fontsize = 20)
plt.xlabel("Total Bill", fontsize=15)
plt.legend()
plt.show()
```
